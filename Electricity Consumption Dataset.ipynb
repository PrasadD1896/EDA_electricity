Electricity is a form of energy that arises from the movement of electrons. It is a fundamental part of modern life and is used in almost every aspect of our daily routine, from powering our homes and workplaces to running the machines and devices we use.

Electricity is generated at power plants and is transmitted to homes, businesses, and industries through power lines. It can be produced using various methods, including fossil fuels, nuclear energy, hydroelectric power, solar power, and wind power.

Electricity is measured in units of watts (W) or kilowatts (kW), and the amount of electricity consumed is typically measured in kilowatt-hours (kWh).

While electricity has revolutionized the modern world, it also poses certain risks, including electrical shock, electrical fires, and electrocution. Therefore, it is important to handle electricity with care and to follow proper safety guidelines when working with electrical systems.

## 4. Analysis of different production sources and consumption of electricity

```python
#Let's see how much electricuty demand we had
plt.figure(figsize = (20,5))
plt.plot(df.index, df['Consumption'],'r')
plt.xlabel('Years')
plt.ylabel('Consumption in MW')
plt.title('Consumption')
plt.show()
```

```python
#Let's see how much electricity produced
plt.figure(figsize=(20,5))
plt.plot(df.index,df['Production'],'b')
plt.title('Power Production')
plt.xlabel('Years')
plt.ylabel('Produced in MW')
plt.show()
```

```python
#Let's compare demand and suplpy
plt.figure(figsize = (20,5))
plt.plot(df.index,df['Production'],'b')
plt.plot(df.index,df['Consumption'],'r')
plt.title('Production vs Consumption')
plt.xlabel('Time')
plt.ylabel('Power in MW')
plt.show()
```

Let's understand how different electricity prodution sources are contributing

```python
columns = {'Nuclear':'orange', 'Wind':'aquamarine', 'Hydroelectric':'b', 'Oil and Gas':'saddlebrown', 'Coal':'black', 'Solar':'y', 'Biomass':'crimson'}
```

```python
for feature in df.columns:
    if feature == 'Consumption' or feature == 'Production':
        pass
    else:
        plt.figure(figsize = (20,5))
        plt.plot(df.index, df[feature], color = columns.get(feature))
        plt.title(feature)
        plt.xlabel('Years')
        plt.ylabel('MW')
        plt.show()
```

Let's split 'DateTime' column into 'Day','Month,'Year','Weekday' for consumption and production analysis

```python
df['Day'] = df.index.day
df['Year'] = df.index.year
df['Month'] = df.index.month
df['WeekDay'] = df.index.day_name()
```

```python
#Yearly consumption and production
df_year = df.groupby('Year').mean()
plt.figure(figsize=(20,5))
plt.plot(df_year.index,df_year['Consumption'],'r')
plt.plot(df_year.index,df_year['Production'],'b')
plt.xlabel('Year')
plt.ylabel('MW')
plt.title('Year wise Consumption vs Production')
plt.legend(['Consumption','Production'])
plt.grid()
plt.show()
```

```python
#Monthly production and consumption
df_Month = df.groupby('Month').mean()
plt.figure(figsize=(20,5))
plt.plot(df_Month.index,df_Month['Consumption'],'r')
plt.plot(df_Month.index,df_Month['Production'],'g')
plt.xlabel('Month of Year')
plt.ylabel('MW')
plt.title('Month wise Consumption vs Production')
plt.legend(['Consumption','Production'])
plt.grid()
plt.show()
```

```python
#Consumption and production on month days
df_day = df.groupby('Day').mean()
plt.figure(figsize=(20,5))
plt.plot(df_day.index,df_day['Consumption'],'r')
plt.plot(df_day.index,df_day['Production'],'g')
plt.xlabel('Days of Month')
plt.ylabel('MW')
plt.title('Daily Consumption vs Production')
plt.legend(['Consumption','Production'])
plt.grid()
plt.show()
```

```python
#Consumption and production on each day
df_WeekDay = df.groupby('WeekDay').mean()
plt.figure(figsize=(20,5))
plt.plot(df_WeekDay.index,df_WeekDay['Consumption'],'r')
plt.plot(df_WeekDay.index,df_WeekDay['Production'],'b')
plt.xlabel('Day of Week')
plt.ylabel('MW')
plt.title('Daily Consumption vs Production')
plt.legend(['Consumption','Production'])
plt.grid()
plt.show()
```

```python
#Energy production by renewable sources of energy
df_renew = df[["Hydroelectric", "Wind", "Solar"]]["2022-06-01 00:00:00" : "2022-08-31 23:59:59"]
plt.figure(figsize=(20,5))
plt.plot(df_renew)
plt.title("Renewable Sources Electricity Production 'MW'")
plt.ylabel('MW')
plt.grid()
plt.show()
```

```python
#Comparison of differernt electricity producing sources
plt.figure(figsize=(25,10))
plt.plot(df_Month.index,df_Month['Nuclear'],'r')
plt.plot(df_Month.index,df_Month['Wind'],'c')
plt.plot(df_Month.index,df_Month['Hydroelectric'],'g')
plt.plot(df_Month.index,df_Month['Oil and Gas'],'orange')
plt.plot(df_Month.index,df_Month['Coal'],'k')
plt.plot(df_Month.index,df_Month['Solar'],'y')
plt.plot(df_Month.index,df_Month['Biomass'],'m')
plt.grid()
plt.legend(['Nuclear','Wind','Hydroelectric','Oil and Gas','Coal','Solar','Biomass'])
plt.show()
```

Let's add two new features to our dataset

```python
df['Renewable'] = df['Wind'] + df['Hydroelectric'] + df['Solar']
df['Non-Renewable'] =df['Nuclear'] + df['Oil and Gas'] + df['Coal'] + df['Biomass']
```

```python
df.groupby('Year')['Renewable'].mean().plot.bar()
```

```python
df.groupby('Year')['Non-Renewable'].mean().plot.bar()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df.groupby('Year')['Renewable','Non-Renewable'].mean())
plt.legend(['Renewable','Non-Renewable'])
```
